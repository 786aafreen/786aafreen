```python
# Function to calculate
# Hamming distance
def hammingDist(str1, str2):
    i = 0
    count = 0
 
    while(i < len(str1)):
        if(str1[i] != str2[i]):
            count += 1
        i += 1
    return count
```

```python

# Driver code 
str1 = "Aafreen"
str2 = "Samreen"
 
# function call
print(hammingDist(str1, str2))
```

```python
s = 'rocket'
s1 = 'rockstar'

print(hammingDist(s, s1))
```
